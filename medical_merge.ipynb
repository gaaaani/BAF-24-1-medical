## EMAIL 기준으로 데이터 병합

```python
train_walk['EMAIL']==train_sleep['EMAIL']
```

```python
comparison_series = train_walk['EMAIL'] == train_sleep['EMAIL']

if False in comparison_series.values:
    print("FALSE")
else:
    print("TRUE")
```

```python
train_data=pd.concat([train_walk,train_sleep],axis=1)
train_data
```

# 결측 열 제거

```python
train_data=train_data.drop(train_data.columns[[4,13,39,42,52]], axis=1)
train_data
```

```python
train_data = train_data.loc[:,~train_data.columns.duplicated()] # 중복되는 EMAIl 열 하나만 남기고 하나만 제거
train_data
```

```python
train_merged_data = pd.merge(train_data, train_think, left_on='EMAIL', right_on='SAMPLE_EMAIL', how='inner')
train_merged_data
```

```python
# 결측치 DOCTOR_NM 열 제거
train_merged_data.drop(['DOCTOR_NM'], axis=1, inplace=True)
train_merged_data
```
